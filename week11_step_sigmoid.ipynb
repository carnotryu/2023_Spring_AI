```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
x = np.arange(-10.0, 10.0, 0.1)
```

## Step function (계단함수)

```python
def step_function(x):
  return np.array(x>0, dtype=np.int)
```

```python
y0 = step_function(x)
```

```python
plt.plot(x, y0)
plt.grid()
plt.show()
```

## Sigmoid
A sigmoid function is a mathematical function having a characteristic "S"-shaped curve or sigmoid curve.

$f(x) = \cfrac{1}{1+e^{-x}}$

```python
def sigmoid(x):
  return 1 / (1 + np.exp(-x))
```

```python
y1 = sigmoid(x)
```

```python
plt.plot(x, y1)
plt.grid()
plt.show()
```

* 일반적인 form



$f(x) = \cfrac{1}{1+e^{-wx}}$

```python
def sigmoid2(w, x):
  return 1 / (1 + np.exp(-w * x))
```

```python
y2 = sigmoid2(1, x)
```

```python
plt.plot(x, y2)
plt.grid()
plt.show()
```

```python
for i in range(10):
  w = 0.2 * (i+1)
  y2 = sigmoid2(w, x)
  plt.plot(x, y2)
  print(w)
plt.grid()
```

* 보다 일반적인 form

$f(x) = \cfrac{1}{1+e^{-w(x-b)}}$

```python
def sigmoid3(w, b, x):
  return 1 / (1 + np.exp(-w* (x - b)))
```

```python
y3 = sigmoid3(1, 1, x)
```

```python
plt.plot(x, y3)
plt.grid()
plt.plot()
```

```python
for i in range(11):
  b = i - 5
  y3 = sigmoid3(1, b, x)
  plt.plot(x, y3)
  plt.plot()
  print(b)
plt.grid()
```
